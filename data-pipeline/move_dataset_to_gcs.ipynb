```python
from google.cloud import storage
import pandas as pd
import os
storage_client = storage.Client(project='ukraine-tweets-381418')
```

```python
already_uploaded = []
def list_blobs(bucket_name):
    bucket = storage_client.bucket(bucket_name)
    """Lists all the blobs in the bucket. https://cloud.google.com/storage/docs/"""
#     blobs = storage_client.list_blobs(bucket_name)
    for blob in bucket.list_blobs(prefix='data'):
        print(str(blob.name))
        already_uploaded.append(str(blob.name).split('data/')[1])
    return already_uploaded
list_blobs("dtc_data_lake_ukraine-tweets-381418")
```

```python
len(already_uploaded)
```

```python

def upload_blob(bucket_name):
    """Uploads a file to the bucket. https://cloud.google.com/storage/docs/ """
    bucket = storage_client.get_bucket(bucket_name)
    for dirname, _, filenames in os.walk('/kaggle/input'):
        for filename in filenames:
            if filename.replace('.gzip', '.gz') in already_uploaded:
                print('already exists')
                continue
            else:
                blob = bucket.blob('delta/'+ filename.replace('.gzip', '.gz'))
                blob.upload_from_filename(os.path.join(dirname, filename))
                print(f"{filename} is uploaded")
upload_blob("dtc_data_lake_ukraine-tweets-381418")
```
